```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
#url for webscrapping
url = 'https://developers.google.com/'
```

```python
#scrape webpage document
webpage = urlopen(url)
```

```python
#create soup object, parse the webpage using html parser
sl_soup = BeautifulSoup(webpage,'html.parser')
```

```python
#close webpage object
webpage.close()
```

```python
#view the content
sl_soup.contents
```

```python
#view output in a well formated way
print(sl_soup.prettify())
```

```python
#view title of scrapped document
print(sl_soup.title)
```

```python
#find all href links present in the webpage
for link in sl_soup.findAll("a",href=True):
    print(link["href"])
```

```python
#view all the headers name
for headers in sl_soup.findAll("h3"):
    print(headers.string)
```
