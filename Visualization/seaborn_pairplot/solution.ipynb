```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
#for viewing plot in jupiter notebook
%matplotlib inline
```

```python
#import dataset
dataset = pd.read_csv('auto_data.csv')
```

```python
#view 1st 5 records
dataset.head()
```

```python
#user defined fuction for origin as origin column contains numeric data
#meaning of the numbers are as follows
# 1 - USA, 2 - Europe, 3- Asia
def origin(num):
    if num == 1:
        return 'USA'
    elif num == 2:
        return 'EUROPE'
    else:
        return 'ASIA'
#use apply function 
dataset['origin'] = dataset['origin'].apply(origin)
```

```python
#view 1st 30 records
dataset.head(30)
```

```python
#draw pairplot using sbn for mpg, weight, origin & with hue origin, set size to 4
# *** hue is variable in dataset to map plot aspects to different colors
sbn.pairplot(dataset[['mpg','weight','origin']],hue = 'origin', size = 4)
```
