```python
#import libraries
import matplotlib.pyplot as plt
#for displaying plot in jupyter
%matplotlib inline
```

```python
#cause of death
cause = 'Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis','Others'
#percentile
percent = [62, 5, 4, 2, 1 , 26]
```

```python
# set pie chart plot properties
# set figure size
plt.figure(figsize = (10,10))
#explode the largest pie 
explode = (0.05,0,0,0,0,0)
#set properties
plt.pie (percent, labels = cause, explode = explode, autopct = '%1.1f%%', startangle = 70)
#set axis to draw pie as circle
plt.axis('equal')
#set title
plt.title('Cause of Death')
#show the plot
plt.show()
```
